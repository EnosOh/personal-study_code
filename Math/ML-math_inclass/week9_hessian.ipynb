```python
import sympy as sp

def calculate_gradient_and_hessian(func_expr, point_to_eval):
    """
    ì£¼ì–´ì§„ í•¨ìˆ˜(func_expr)ì˜ ê·¸ë˜ë””ì–¸íŠ¸ì™€ í—¤ì‹œì•ˆ í–‰ë ¬ì„ ê³„ì‚°í•˜ê³ ,
    ì£¼ì–´ì§„ ì§€ì (point_to_eval)ì—ì„œ ê·¸ ê°’ì„ í‰ê°€í•˜ì—¬ ì¶œë ¥í•©ë‹ˆë‹¤.

    Args:
        func_expr (sympy.Expr): ì‹¬íŒŒì´ë¡œ ì •ì˜ëœ í•¨ìˆ˜ í‘œí˜„ì‹
        point_to_eval (dict): í‰ê°€í•  ì§€ì . ì˜ˆ: {x: 1, y: 2}
    """
    # í•¨ìˆ˜ì— ì‚¬ìš©ëœ ë³€ìˆ˜ë“¤ì„ ìë™ìœ¼ë¡œ ì°¾ì•„ ì •ë ¬
    variables = sorted(list(func_expr.free_symbols), key=str)
    
    print("===================================================")
    print(f"âœ… {len(variables)}ë³€ìˆ˜ í•¨ìˆ˜ì— ëŒ€í•œ ê³„ì‚° ì‹œì‘")
    print("===================================================")
    
    # Step 0: ì •ì˜ëœ í•¨ìˆ˜ ì¶œë ¥
    print("\nğŸ“Œ Step 0: ì •ì˜ëœ í•¨ìˆ˜")
    sp.pprint(func_expr)
    
    # Step 1: ê·¸ë˜ë””ì–¸íŠ¸ ë²¡í„° ê³„ì‚° (1ì°¨ í¸ë¯¸ë¶„)
    gradient_vector = sp.Matrix([sp.diff(func_expr, var) for var in variables])
    print("\nğŸ“Œ Step 1: ê·¸ë˜ë””ì–¸íŠ¸ ë²¡í„° (âˆ‡f)")
    sp.pprint(gradient_vector)
    
    # Step 2: í—¤ì‹œì•ˆ í–‰ë ¬ ê³„ì‚° (2ì°¨ í¸ë¯¸ë¶„)
    # sympyì˜ ë‚´ì¥ hessian í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ í•œ ë²ˆì— ê³„ì‚°
    hessian_matrix = sp.hessian(func_expr, variables)
    print("\nğŸ“Œ Step 2: í—¤ì‹œì•ˆ í–‰ë ¬ (H)")
    sp.pprint(hessian_matrix)

    # Step 3: íŠ¹ì • ì§€ì ì—ì„œì˜ ìˆ˜ì¹˜ í‰ê°€
    grad_val = gradient_vector.subs(point_to_eval)
    hessian_val = hessian_matrix.subs(point_to_eval)
    
    # í‰ê°€ ì§€ì ì„ ë¬¸ìì—´ë¡œ ì˜ˆì˜ê²Œ ë§Œë“¤ê¸°
    point_str = ", ".join([f"{k}={v}" for k, v in point_to_eval.items()])
    
    print(f"\nğŸ“Œ Step 3: ({point_str}) ì—ì„œì˜ í‰ê°€")
    print("\n[í‰ê°€ ê²°ê³¼] ê·¸ë˜ë””ì–¸íŠ¸ ë²¡í„°:")
    sp.pprint(grad_val)
    print("\n[í‰ê°€ ê²°ê³¼] í—¤ì‹œì•ˆ í–‰ë ¬:")
    sp.pprint(hessian_val)
    print("\n\n")


# --- ì´ˆê¸° ì„¤ì •: ì‚¬ìš©í•  ìˆ˜ ìˆëŠ” ë³€ìˆ˜ë“¤ì„ ë¯¸ë¦¬ ì •ì˜ ---
x, y, z = sp.symbols('x y z')


# --- ì˜ˆì œ 1: 2ë³€ìˆ˜ í•¨ìˆ˜ (ê¸°ì¡´ ì˜ˆì œ) ---
# ì´ ë¶€ë¶„ì˜ í•¨ìˆ˜ ì •ì˜ì™€ í‰ê°€ ì§€ì ë§Œ ìˆ˜ì •í•˜ì—¬ ì‚¬ìš©í•˜ì„¸ìš”.
func_2d =  x**3 - 3*x**2*y + 3*x*y**2 - y**3
point_2d = {x: 1, y: 2}
calculate_gradient_and_hessian(func_2d, point_2d)


# --- ì˜ˆì œ 2: 3ë³€ìˆ˜ í•¨ìˆ˜ ---
# 3ë³€ìˆ˜ í•¨ìˆ˜ë¡œ ë°”ê¿”ì„œ í…ŒìŠ¤íŠ¸
func_3d = x**2 * y + y**2 * z + z**2 * x
point_3d = {x: 1, y: 2, z: 3}
calculate_gradient_and_hessian(func_3d, point_3d)
```
