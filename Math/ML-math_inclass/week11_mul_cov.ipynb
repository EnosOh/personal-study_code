```python
from sympy import Matrix, Rational, pprint

# 1. 데이터 정의
X_samples = [Matrix([1,2]), Matrix([3, 4])]
Y_samples = [Matrix([1, 3]), Matrix([2, 4])]
n = len(X_samples)

# 2. 평균 벡터 계산 (초깃값 추가!)
x_bar = sum(X_samples, Matrix.zeros(2, 1)) / n
y_bar = sum(Y_samples, Matrix.zeros(2, 1)) / n

# 3. E[xy^T] 계산
xy_outer_sum = sum([x * y.T for x, y in zip(X_samples, Y_samples)], Matrix.zeros(2, 2))
E_xyT = xy_outer_sum / n

# 4. E[x] * E[y]^T 계산
Ex_EyT = x_bar * y_bar.T

# 5. 공분산 행렬 계산
Cov_xy = E_xyT - Ex_EyT

# 출력
print("E[x]:")
pprint(x_bar)
print("\nE[y]:")
pprint(y_bar)
print("\nE[xy^T]:")
pprint(E_xyT)
print("\nE[x] * E[y]^T:")
pprint(Ex_EyT)
print("\nCov[x, y]:")
pprint(Cov_xy)
```

### 표본 공분산 같은 data 두개로 하는거 cov(x,x)

```python
from sympy import Matrix, pprint

X_samples = [Matrix([1, 2]), Matrix([3, 4])]
n = len(X_samples)
x_bar = sum(X_samples, Matrix.zeros(2, 1)) / n

# Outer product sum
S = sum([(x - x_bar) * (x - x_bar).T for x in X_samples], Matrix.zeros(2,2))
cov = S / n

pprint(cov)
```
