```python
import sympy as sp

# Step 0. ê¸°í˜¸ ì •ì˜ ë° í•¨ìˆ˜ ì„¤ì •
x, y = sp.symbols('x y')
f = x**2 + 2*x*y + y**3
x0, y0 = 1, 2  # ê¸°ì¤€ì 

dx = x - x0
dy = y - y0

print("ğŸ“Œ Step 0: í•¨ìˆ˜ ì •ì˜")
sp.pprint(f)
print(f"ê¸°ì¤€ì : (x0, y0) = ({x0}, {y0})")
print(f"Î´x = x - {x0} = {dx}, Î´y = y - {y0} = {dy}")

# Step 1. Gradient ê³„ì‚°
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
grad = [df_dx.subs({x: x0, y: y0}), df_dy.subs({x: x0, y: y0})]

print("\nğŸ“Œ Step 1: Gradient")
print("âˆ‚f/âˆ‚x =", df_dx, "â†’", grad[0])
print("âˆ‚f/âˆ‚y =", df_dy, "â†’", grad[1])

# Step 2. Hessian ê³„ì‚°
d2f_xx = sp.diff(df_dx, x)
d2f_xy = sp.diff(df_dx, y)
d2f_yy = sp.diff(df_dy, y)

H = sp.Matrix([[d2f_xx, d2f_xy], [d2f_xy, d2f_yy]])
H_val = H.subs({x: x0, y: y0})

print("\nğŸ“Œ Step 2: Hessian")
sp.pprint(H)
print("â†’ í‰ê°€ at (1,2):")
sp.pprint(H_val)

# Step 3. 3ì°¨ ë„í•¨ìˆ˜ ëª¨ë‘ ê³„ì‚°
d3f_xxx = sp.diff(d2f_xx, x)
d3f_xxy = sp.diff(d2f_xx, y)
d3f_xyy = sp.diff(d2f_xy, y)
d3f_yyy = sp.diff(d2f_yy, y)

print("\nğŸ“Œ Step 3: 3ì°¨ ë„í•¨ìˆ˜")
print("âˆ‚Â³f/âˆ‚xÂ³ =", d3f_xxx)
print("âˆ‚Â³f/âˆ‚xÂ²âˆ‚y =", d3f_xxy)
print("âˆ‚Â³f/âˆ‚xâˆ‚yÂ² =", d3f_xyy)
print("âˆ‚Â³f/âˆ‚yÂ³ =", d3f_yyy)

# Step 4. ê° í…Œì¼ëŸ¬ í•­ ê³„ì‚°
T0 = f.subs({x: x0, y: y0})

T1 = grad[0]*dx + grad[1]*dy

T2 = (1/2) * (dx*dx*H_val[0,0] + 2*dx*dy*H_val[0,1] + dy*dy*H_val[1,1])

T3 = (1/6) * d3f_yyy.subs({x: x0, y: y0}) * dy**3  # ë‚˜ë¨¸ì§€ í•­ì€ 0

print("\nğŸ“Œ Step 4: Taylor í•­ ê³„ì‚°")
print(f"Tâ‚€ = {T0}")
print(f"Tâ‚ = {T1}")
print(f"Tâ‚‚ = {T2}")
print(f"Tâ‚ƒ = {T3}")

# Step 5. ì „ì²´ í…Œì¼ëŸ¬ ì „ê°œ í•©
T_total = T0 + T1 + T2 + T3

print("\nâœ… ìµœì¢… 3ì°¨ í…Œì¼ëŸ¬ ë‹¤í•­ì‹:")
sp.pprint(sp.simplify(T_total))
```
