```python
import numpy as np

x = np.array([2, 4, 6, 8])
y = np.array([1, 3, 2, 5])

# 공분산 행렬 계산
cov_matrix = np.cov(x, y,bias=True)  # 2x2 행렬: cov(x,x), cov(x,y), cov(y,x), cov(y,y)
print("공분산 행렬:\n", cov_matrix)

# 공분산 값만 출력 (x와 y의 공분산)
print("공분산:", cov_matrix[0, 1])
```

```python
import sympy as sp

# --- 데이터 정의 ---
x_data = [2, 4, 6, 8]
y_data = [1, 3, 2, 5]

print("--- 입력 데이터 ---")
print(f"x = {x_data}")
print(f"y = {y_data}\n")

# --- 1. 평균 계산 ---
print("--- 1. 평균 계산 ---")
n = len(x_data)
mean_x = sp.Rational(sum(x_data), n)
mean_y = sp.Rational(sum(y_data), n)
print(f"x의 평균 (정확한 값): {mean_x}")
print(f"y의 평균 (정확한 값): {mean_y}")
print(f"y의 평균 (소수점): {mean_y.evalf()}\n")

# --- 2. 모집단 공분산 Cov(x, y) 계산 ---
print("--- 2. 모집단 공분산 계산 ---")
sum_of_products = sum((xi - mean_x) * (yi - mean_y) for xi, yi in zip(x_data, y_data))
cov_xy = sum_of_products / n
print(f"x와 y의 공분산 (정확한 값): {cov_xy}")
print(f"x와 y의 공분산 (소수점): {cov_xy.evalf()}\n")

# --- 3. 모집단 분산 계산 ---
print("--- 3. 모집단 분산 계산 ---")
var_x = sum((xi - mean_x)**2 for xi in x_data) / n
print(f"x의 분산 (정확한 값): {var_x}")
print(f"x의 분산 (소수점): {var_x.evalf()}")

var_y = sum((yi - mean_y)**2 for yi in y_data) / n
print(f"y의 분산 (정확한 값): {var_y}")
print(f"y의 분산 (소수점): {var_y.evalf()}\n")

# --- 4. 공분산 행렬 구성 ---
print("--- 4. 공분산 행렬 (정확한 값) ---")
covariance_matrix = sp.Matrix([
    [var_x, cov_xy],
    [cov_xy, var_y]
])
sp.pprint(covariance_matrix)
print("\n--- 공분산 행렬 (소수점) ---")
sp.pprint(covariance_matrix.evalf())
print()

# --- 5. 상관계수 계산 ---
print("--- 5. 상관계수 계산 ---")
std_dev_x = sp.sqrt(var_x)
std_dev_y = sp.sqrt(var_y)
correlation_coefficient = cov_xy / (std_dev_x * std_dev_y)

print("x의 표준편차 (정확한 값):")
sp.pprint(std_dev_x)
print(f"x의 표준편차 (소수점): {std_dev_x.evalf()}\n")

print("y의 표준편차 (정확한 값):")
sp.pprint(std_dev_y)
print(f"y의 표준편차 (소수점): {std_dev_y.evalf()}\n")

print("x와 y의 상관계수 (정확한 값):")
sp.pprint(correlation_coefficient)
print("\n상관계수 (소수점):")
sp.pprint(correlation_coefficient.evalf())
```
